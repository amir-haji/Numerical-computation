            الی که به تازگی در مدرسه‌ی فدرال درس محاسبات عددی را گذرانده تصمیم می‌گیرد تا جمعیت زامبی‌ها و انسان‌ها را مدل کند.
    </div>
    <div>
جمعیت انسان‌ها را با
        \(h\)
        و جمعیت زامبی‌ها را با
        \(z\)
        نشان می‌دهیم. در حالت ساده معادله‌ی دیفرانسیل سیستم این شکلی است:
    </div>
    <p>
        \begin{aligned}
        {\frac {dh}{dt}} &= \alpha h-\beta hz,\\
        {\frac {dz}{dt}} &= \delta hz-\gamma z,
        \end{aligned}
    </p>
    <div>
                 به این معادلات مرتبه‌ی اول ساده معادلات Lotka–Volterra یا شکار و شکارچی نیز می‌گویند که تغییر جمعیت شکار و شکارچی (اینجا انسان و زامبی) را مدل می‌کند. در صورت علاقه می‌توانیم علت هر جمله در معادله را بررسی کنیم:
    <ul>
        <li>$\alpha h$: زاد و ولد انسان‌ها منهای مرگ و میر آنها که متناسب با جمعیتشان است.</li>
        <li>$\beta hz$: زامبی شدن انسان‌ها که به دلیل واگیردار بودن متناسب با جمعیت هر دو گونه است.</li>
        <li>$\delta hz$: زامبی شدن انسان‌ها منهای کشتن زامبی‌ها توسط انسان‌ها که باز متناسب با جمعیت هر دو گونه است.</li>
        <li>$\gamma z$: مرگ طبیعی زامبی‌ها.</li>
    </ul>
    اما در دنیای آخرالزمانی منابع زندگی به شدت محدود است بنابر این انسان‌ها حتی در نبود زامبی‌ها هم رشد جمعیت طبیعی ندارد و با زیاد شدن جمعیت رشد آن‌ها محدود می‌شود:
    </div>
    <p>
        \begin{aligned}
        {\frac {dh}{dt}} &= \alpha h_0 \tanh{\frac{h}{h_0}}-\beta hz,\\
        {\frac {dz}{dt}} &= \delta hz-\gamma z
        \end{aligned}
    </p>
        که در صورتی که با خواص تابع \(\tanh\) آشنا باشید می‌دانید که در \(h\) های بزرگ جمله‌ی زاد و ولد به \(\alpha h_0\) میل می‌کند و در \(h\) های کوچک به صورت خطی بعنی \(\alpha h\) عمل می‌کند (مثل معادلات قبل).
    <p>
    </p>
    <div>
        (۱۰ امتیاز) تابع human_zombie_dynamic را طوری تعریف کنید که با گرفتن زمان (که اینجا تابعیتی از آن نداریم و در تابع استفاده نمی‌شود)، جمعیت زامبی‌ها و انسان‌ها، مشتق جمعیت انسان‌ها و مشتق جمعیت زامبی‌ها را به ترتیب برگرداند. در واقع کل معادله‌ی دیفرانسیل را در این تابع می‌گنجانیم.
        نکته‌ی مهم: حتما از شکل آخرالزمانی معادلات استفاده کنید و ضمنا می‌توانید از تابع np.tanh نیز استفاده کنید.
    </div>
    <b>در تمام نوت‌بوک می‌توانید توابع و فرمت پیشنهادی کد را تغییر دهید و به روش خودتان پیاده‌سازی کنید تا جایی که نتیجه درست باشد مسئله‌ای نیست.</b>
</div>

```python
import numpy as np

def human_zombie_dynamic(t, h, z, alpha, beta, delta, gamma, h0):
    dhdt = alpha * h0 * np.tanh(h / h0) - beta * h * z  #: Your code here ##
    dzdt = delta * h * z - gamma * z  #: Your code here ##

    return np.array([dhdt, dzdt])

def dynamic_generator(alpha, beta, delta, gamma, h0):
    return lambda t, x: human_zombie_dynamic(t, x[0], x[1], alpha, beta, delta, gamma, h0)
```

    دارد که در آن مقادیر جمعیت انسان‌ها و زامبی‌ها قرار دارد. از این تابع ساده شده برای حل عددی معادله‌ی دیفرانسیل جلوتر استفاده می‌کنیم.
<h1>رانگ کوتا(۶۰ امتیاز)</h1>
می‌خواهیم از روش رانگ کوتای مرتبه‌ی ۴ برای حل عددی دستگاه معادلات دیفرانسیل بالا استفاده کنیم. معادلات همان شکل آشنای قبلی را دارند با این تفاوت که
    x
    ها دیگر بردار هستند و عدد نیستند. البته معادلات به همان شکل قبلی برقرار است. برای معادله دیفرانسیل
  $x' = f(t, x)$:
$$
\begin{align*}
k_1 &= dt \cdot f(t_n, x_n) \\
k_2 &= dt \cdot f\left(t_n + \frac{dt}{2}, x_n + \frac{1}{2} \cdot k_1\right) \\
k_3 &= dt \cdot f\left(t_n + \frac{dt}{2}, x_n + \frac{1}{2} \cdot k_2\right) \\
k_4 &= dt \cdot f(t_n + dt, x_n + k_3) \\
x_{n+1} &= x_n + \frac{1}{6} \cdot (k_1 + 2k_2 + 2k_3 + k_4)
\end{align*}
$$
    حال تابع زیر را کامل کنید تا با گرفتن تابع f و نقاط شروع، پایان و فاصله‌ی زمانی معادله دیفرانسیل
  $x' = f(t, x)$
    که x یک بردار است را در این بازه حل کند.
</div>

```python
def runge_kutta(f, x0, t0, tf, dt):
    """
    Solve a system of differential equations using the Runge-Kutta method.

    Parameters:
        f: function
            The function defining the system of differential equations.
            It should accept two arguments: t (time) and x (state vector).
        x0: array-like
            Initial state vector.
        t0: float
            Initial time.
        tf: float
            Final time.
        dt: float
            Step size.

    Returns:
        t_values: array
            Array of time values.
        x_values: array
            Array of state vectors corresponding to the time values.
    """
    t_values = np.arange(t0, tf + dt, dt)
    num_steps = len(t_values)
    dim = len(x0)
    x_values = np.zeros((num_steps, dim))
    x_values[0] = x0


    for i in range(1, num_steps):
        x_i = x_values[i - 1]
        t_i = t_values[i - 1]
        
        k1 = dt * f(t_i, x_i)
        k2 = dt * f(t_i + dt/2, x_i + k1/2)
        k3 = dt * f(t_i + dt/2, x_i + k2/2)
        k4 = dt * f(t_i + dt, x_i + k3)
        
        x_values[i] = x_i + (k1 + k2 + k3 + k4) / 6

    return t_values, x_values
```

```python
import matplotlib.pyplot as plt

alpha = 0.1
beta = 0.0011
delta = 0.001
gamma = 0.1
h0 = 100

x0 = np.array([100, 100])
t0 = 0
tf = 500
dt = 0.1

f1 = dynamic_generator(alpha, beta, delta, gamma, h0)

def plot_population(f, x0, t0, tf, dt):
    t_values, x_values = runge_kutta(f, x0, t0, tf, dt)
    fig = plt.figure(figsize = (1 * 10, 2 * 10))
    
    plt.subplot(2, 1, 1)
    plt.plot(t_values, x_values[:, 0], label = 'Humans')
    plt.plot(t_values, x_values[:, 1], label = 'Zombies')
    plt.ylabel('Population')
    plt.xlabel('Time')
    plt.legend(loc = 'best')
    
    plt.subplot(2, 1, 2)
    plt.scatter(x_values[:, 0], x_values[:, 1])
    plt.ylabel('Zombies')
    plt.xlabel('Humans')

plot_population(f1, x0, t0, tf, dt)
```

<h2>نقطه‌ی تعادل(۳۰ امتیاز)</h2>
حتما در نمودارهای بالا متوجه شدید که سیستم به نقطه‌ای میل کرده است.
نقطه‌ی تعادل جایی نیست که زامبی‌ها و انسان‌ها با هم به خوبی و خوشی زندگی می‌کنند، بلکه از نظر ما جایی است که جمعیت زامبی‌ها و انسان‌ها تغییر نمی‌کند. (البته لزوما سیستم به همه‌ی نقاط تعادل میل نمی‌کند و ممکن است «پایدار» نباشند.) حال می‌خواهیم این نقطه را پیدا کنیم. در این نقطه:
\begin{aligned}{\frac {dh}{dt}}&=\alpha h_0 \tanh{\frac{h^*}{h_0}}-\beta h^*z^* = 0,\\{\frac {dz}{dt}}&=\delta h^*z^*-\gamma z^* = 0\end{aligned}
اما بر خلاف معادلات مدل خطی به دلیل وجود عامل
tanh
این معادلات را برای مدل ما نمی‌توان به صورت تحلیلی حل کرد. برای پیدا کردن نقطه‌ی تعادلی باید از روش‌های عددی استفاده کنیم.
<h3>نیوتن رافسون</h3>
یکی از این روش‌ها که با آن آشنا هستید روش نیوتن رافسون است. در این روش از یک نقطه شروع می‌کنیم و هر بار نقطه‌ی بعدی از این رابطه بدست می‌آید:
$$
\mathbf{x}_{n+1} = \mathbf{x}_n - J^{-1}_{f}(\mathbf{x}_n) \cdot f(\mathbf{x}_n)
$$
که در آن
$ \mathbf{x}_n $
بردار متغییرها در مرحله‌ی nام است
$J$ ماتریس جاکوبی است و
و $f$ همان بردار تابع ماست.
<div>
    حال با استفاده از این روش و تکمیل تابع زیر نقطه‌ی تعادل را بیابید.
    از کتابخانه‌ی نامپای برای یافتن وارون ماتریس می توانید استفاده کنید. (دقت کنید که جاکوبین ماتریس به صورت تحلیلی قابل محاسبه است و از متغییرها می‌توانید داخل تابع استفاده کنید.)
نقطه‌ی شروع را می‌توانید بر اساس نمودارهایی که بالا بدست آوردید انتخاب کنید.
</div>
</div>

```python
from numpy.linalg import inv

alpha = 0.1
beta = 0.0011
delta = 0.001
gamma = 0.1
h0 = 100

f1 = dynamic_generator(alpha, beta, delta, gamma, h0)

def newton_raphson(f, x0, iter=100):
    x = x0
    for _ in range(iter):
        f_x = f(0, x)  # Time is not used in the function
        h, z = x
        H = np.array([[alpha * (1 - np.tanh(h / h0)) - beta * z, - beta * h], [delta * z, delta * h - gamma]])
        J = inv(H)
        
        x = x - np.dot(J, f_x)

    return x

newton_raphson(f1, np.array([100, 100]))  # Change the starting point if you wish
```
